## This notebook is used to update the deployments feature class from the master deployments spreadsheet.


#### Run this cell to connect to your GIS and get started:

```python
import pandas as pd
import numpy as np
import requests
import io
import datetime
from arcgis.gis import GIS
from copy import deepcopy 
```

```python
gis = GIS("home")
deploymentsLyr = gis.content.get('69ff90b5ceed4fb8be0cafb531a7a2c7').layers[0]
```

```python
url="https://docs.google.com/spreadsheets/d/e/2PACX-1vT64Lj-s6fPoGctFUhscbi6XTEYSA3vR0sf__FGBUW363erf2V5O2LNPTJdGUGpQvve81-cyoBdGzsJ/pub?gid=1494601508&single=true&output=csv"
s=requests.get(url).content
df=pd.read_csv(io.StringIO(s.decode('utf-8')))
df = df.replace(np.nan, '', regex=True)
display(df)
```

```python
deploymentsAdds=[]
for index, row in df.iterrows():
    f = {"geometry":{"x": None, "y": None},"attributes":{"title": None, "description": None, "disastertype": None, "mdr_link": None, "startdate": None}}
    # process the gemoetry
    f['geometry']['x'], f['geometry']['y'] = float(row['Longitude']),float(row['Latitude'])    
    f['attributes']["title"] = row['Title']
    f['attributes']["description"] = row['Description']
    f['attributes']["disastertype"] = row['Disaster Type']
    f['attributes']["mdr_link"] = row['mdrURL']
    f['attributes']["startdate"] = datetime.datetime.strptime(row['Start Date'], '%Y/%m/%d')
    f['attributes']["partners"] = str(row['Partner(s)'])
    f['attributes']["ActivityCategory"] = row['Activity Category']
    deploymentsAdds.append(deepcopy(f))

print(deploymentsAdds[0])
```

```python
deploymentsLyr.manager.truncate()
deploymentsLyr.edit_features(adds=deploymentsAdds)
```
